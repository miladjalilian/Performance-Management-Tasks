```python
data = pd.merge(LTE_KPI,Connected,how='left',on=['Datetime','eNodeB'])
```

```python
data['Sector'] = data['eNodeB']
data.Sector = data.eNodeB.str.slice(0, 2) + data.eNodeB.str.slice(4, 9)

data_den = data.groupby(['Index','Sector', 'Datetime'])['Connected_User'].agg(sum)

data_mult = pd.DataFrame()
data_mult['UDT'] = data['User_Thr_Mbps'] * data['Connected_User']
data_mult['Datetime'] = data['Datetime']
data_mult['Sector'] = data['Sector']
data_mult['Index'] = data['Index']

data_mult_sum = data_mult.groupby(['Index','Sector', 'Datetime'])['UDT'].agg(sum)

data_merged = pd.concat([ data_mult_sum, data_den], axis = 1, sort=False)
data_merged['User_Thr_Mbps'] = data_merged['UDT'] / (data_merged['Connected_User'] + (data_merged['Connected_User'] == 0))
data_merged.drop(['UDT'], axis=1, inplace = True)
data_merged_df = data_merged.reset_index()

df1argmax = data_merged_df.groupby('Sector', as_index=False).apply(lambda x: x.loc[x.Connected_User.idxmax(),])
df1argmax['Site'] = df1argmax.Sector.str.slice(0, 6)
```
