```python
M=[]
for i in range(len(df)): # in each row
    x = df.loc[df.index[i]].sort_values()
    s = df.index[i]
    for j in range(1,4): # in each column
        if x[j]<35000:
            s = s +'|'+x.index[j]+'|'+str(round(x[j]/1000,2))
    M.append(s)
```

```python
df6=pd.DataFrame(data=M)
```

```python
df4=[]
for i in range(len(df6)):
    df4.append(df6[0][i].split("|"))
```

```python
df4_3=pd.DataFrame(df4,columns=['SiteName','1st Tier','Distance1','2nd Tier','Distance2','3rd Tier','Distance43'])
```

```python
df4_3
```

```python
df4_3['Tier1']=round(df4_3['Distance1'].astype(np.float),1)
df4_3['Tier2']=round(df4_3['Distance2'].astype(np.float),1)
df4_3['Tier3']=round(df4_3['Distance43'].astype(np.float),1)
```

```python
df4_3.to_excel('Tier_Dashboard.xlsx')
```
