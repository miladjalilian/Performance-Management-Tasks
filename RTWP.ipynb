```python
excel=[]
for file in os.listdir('D:\Saeid Lte Congested 30 March'):
    #print(file)
    if file[8:12] == 'RTWP':
        excel.append(file)
```

```python
import pandas as pd
import numpy as np

df_RTWP = []

df_RTWP = pd.concat([pd.read_excel('D:\Saeid Lte Congested 30 March\\'+ n,'Average RTWP dbm') for n in excel])
```

```python
df_RTWP['Site'] = df_RTWP['ElementID1'].str[0:2] + df_RTWP['ElementID1'].str[4:8]
```

```python
NeDB = pd.read_excel(r'D:\Saeid Lte Congested 30 March\Network Site DB.xlsx', sheet_name='Main DB')
```

```python
NeDB.columns
```

```python
df_RTWP = df_RTWP.merge(NeDB[['Site','City','City (Farsi)']], how = 'left', on = 'Site')
```

```python
df_RTWP_SZ = []
df_RTWP_SZ = df_RTWP[df_RTWP['City_y'] == 'Shiraz']
```

```python
Normal = df_RTWP_SZ[df_RTWP_SZ['Average RTWP dbm']<=-95].groupby('ElementID1',as_index=False)['Average RTWP dbm'].count()
Normal.rename(columns={'Average RTWP dbm':'Normal'},inplace=True)

Major = df_RTWP_SZ[(df_RTWP_SZ['Average RTWP dbm']<=-85) & (df_RTWP_SZ['Average RTWP dbm']>-95)].groupby('ElementID1',as_index=False)['Average RTWP dbm'].count()
Major.rename(columns={'Average RTWP dbm':'Major'},inplace=True)

Critical = df_RTWP_SZ[df_RTWP_SZ['Average RTWP dbm']>-85].groupby('ElementID1',as_index=False)['Average RTWP dbm'].count()
Critical.rename(columns={'Average RTWP dbm':'Critical'},inplace=True)
```

```python
df_RTWP_SZ = df_RTWP_SZ.merge(Normal, how='left', on='ElementID1')
df_RTWP_SZ = df_RTWP_SZ.merge(Major, how='left', on='ElementID1')
df_RTWP_SZ = df_RTWP_SZ.merge(Critical, how='left', on='ElementID1')
```

```python
df_RTWP_SZ['Frequency Band'] = np.where(df_RTWP_SZ['ElementID1'].str[2:4]=='2U', '900MHz', '2100MHz')
```

```python
celllist_RD = []
celllist_RD = df_RTWP_SZ[['ElementID','ElementID1','City_y','Normal','Major','Critical','Frequency Band']]
celllist_RD.drop_duplicates(inplace=True)
```

```python
celllist_RD
```

```python
writer = pd.ExcelWriter(r'D:\Saeid Lte Congested 30 March\RTWP_WorstCells2.xlsx', engine='xlsxwriter')

celllist_RD.to_excel(writer,sheet_name='Cell_List')
writer.save()
```
